```python
modulo7/
│
├── aula4_questao1.py
├── aula4_questao2.py
├── aula4_questao3.py
├── aula4_questao4.py
├── aula4_questao5.py
└── aula4_questao6.py

modulo7/aula4_questao1.py

import os

frase = input("Digite uma frase: ")
caminho_arquivo = os.path.join(os.getcwd(), "frase.txt")

with open(caminho_arquivo, "w", encoding="utf-8") as arquivo:
    arquivo.write(frase)

print(f"Frase salva em {caminho_arquivo}")

import re

with open("frase.txt", "r", encoding="utf-8") as arquivo:
    conteudo = arquivo.read()

palavras = re.findall(r'\b\w+\b', conteudo, flags=re.UNICODE)

with open("palavras.txt", "w", encoding="utf-8") as arquivo:
    for palavra in palavras:
        arquivo.write(palavra + "\n")

with open("palavras.txt", "r", encoding="utf-8") as arquivo:
    print(arquivo.read())


modulo7/aula4_questao2.py

import re

with open("frase.txt", "r", encoding="utf-8") as arquivo:
    conteudo = arquivo.read()

palavras = re.findall(r'\b\w+\b', conteudo, flags=re.UNICODE)

with open("palavras.txt", "w", encoding="utf-8") as arquivo:
    for palavra in palavras:
        arquivo.write(palavra + "\n")

with open("palavras.txt", "r", encoding="utf-8") as arquivo:
    print(arquivo.read())

    modulo7/aula4_questao3.py

    import re

with open("estomago.txt", "r", encoding="utf-8") as arquivo:
    linhas = arquivo.readlines()

print("Primeiras 25 linhas:")
print("".join(linhas[:25]))

print(f"\nNúmero de linhas do arquivo: {len(linhas)}")

linha_maior = max(linhas, key=len)
print(f"\nLinha com maior número de caracteres:\n{linha_maior}")

texto = "".join(linhas).lower()
nonato_count = len(re.findall(r'\bnonato\b', texto))
iria_count = len(re.findall(r'\bíria\b', texto))

print(f"\nMenções de 'Nonato': {nonato_count}")
print(f"Menções de 'Íria': {iria_count}")

modulo7/aula4_questao4.py

import random

with open("gabarito_forca.txt", "r", encoding="utf-8") as arquivo:
    palavras = [linha.strip() for linha in arquivo if linha.strip()]

palavra = random.choice(palavras).upper()
letras_adivinhadas = ["_" for _ in palavra]
tentativas = 0
max_tentativas = 6

with open("gabarito_enforcado.txt", "r", encoding="utf-8") as arquivo:
    estagios = arquivo.read().split("=========\n")

def imprime_enforcado(tentativas):
    print(estagios[tentativas])

while tentativas < max_tentativas and "_" in letras_adivinhadas:
    print("Palavra: " + " ".join(letras_adivinhadas))
    letra = input("Digite uma letra: ").upper()
    if letra in palavra:
        for i, l in enumerate(palavra):
            if l == letra:
                letras_adivinhadas[i] = letra
    else:
        tentativas += 1
        imprime_enforcado(tentativas-1)

if "_" not in letras_adivinhadas:
    print("Parabéns! Você acertou a palavra:", palavra)
else:
    print("Fim de jogo! A palavra era:", palavra)


modulo7/aula4_questao5.py


livros = [
    ["O Caçador de Pipas", "Khaled Hosseini", 2003, 368],
    ["Torto Arado", "Itamar Vieira Junior", 2019, 264],
    ["Dom Casmurro", "Machado de Assis", 1899, 256],
    ["1984", "George Orwell", 1949, 328],
    ["O Alquimista", "Paulo Coelho", 1988, 208],
    ["Grande Sertão: Veredas", "João Guimarães Rosa", 1956, 624],
    ["Cem Anos de Solidão", "Gabriel García Márquez", 1967, 432],
    ["A Menina que Roubava Livros", "Markus Zusak", 2005, 480],
    ["O Hobbit", "J.R.R. Tolkien", 1937, 310],
    ["Sapiens", "Yuval Noah Harari", 2011, 498]
]

with open("meus_livros.csv", "w", encoding="utf-8") as arquivo:
    arquivo.write("Título,Autor,Ano de publicação,Número de páginas\n")
    for livro in livros:
        arquivo.write(",".join(map(str, livro)) + "\n")

modulo7/aula4_questao6.py

import csv

mais_populares = []

with open("spotify-2023.csv", "r", encoding="latin-1") as arquivo:
    leitor = csv.reader(arquivo)
    next(leitor)  # Pula cabeçalho
    musicas_anos = {}

    for linha in leitor:
        if '"' in linha[0] or '"' in linha[1]:  # Ignora músicas problemáticas
            continue
        track_name = linha[0]
        artist_name = linha[1]
        released_year = int(linha[3])
        streams = int(linha[8])

        if 2012 <= released_year <= 2022:
            if released_year not in musicas_anos or streams > musicas_anos[released_year][3]:
                musicas_anos[released_year] = [track_name, artist_name, released_year, streams]

anos_ordenados = sorted(musicas_anos.keys())
lista_final = [musicas_anos[ano] for ano in anos_ordenados]

print(lista_final)

```
